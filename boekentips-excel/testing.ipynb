```python
import requests

for i in data.itertuples():
    author = i.schrijver
    title = i.titel
    language = i.taal

    if language == 'Nederlands':
        language = 'dut'
    elif language == 'Engels':
        language = 'eng'

    url = 'https://openlibrary.org/search.json'

    # Fetching data from Open Library API
    x = requests.get(url, params={'author': author, 'title': title, 'language': language})
    results = x.json().get('numFound')

    # Check if the book exists in the specified language
    if results <= 0:
        # Check if the book exists in other languages,
        # does not work at all times for dutch books since the english title is not always the same or similar
        y = requests.get(url, params={'author': author, 'title': title})
        results = y.json().get('numFound')

        if results <= 0:
            print(f"Not found: {author} - {title} in the language: {language}")
            print("need to add this book to the database at the url:")
            print("https://openlibrary.org/books/add")
            print(f"isbn: {i.isbn}, naam: {i.titel}, schrijver: {i.schrijver}, taal: {i.taal}")
        else:
            print(f"Not found: {author} - {title} in the language: {language}")
            print("Does exist in other languages, work can be found at:")
            print("https://openlibrary.org" + y.json().get('docs')[0].get('key'))
            print(f"isbn: {i.isbn}, naam: {i.titel}, schrijver: {i.schrijver}, taal: {i.taal}")
    else:
        print(f"Found {results} results for: {author} - {title} in the language: {language}")
        print("You can find the book at the url:")
        print(x.url)
        
    print("\n")
```
